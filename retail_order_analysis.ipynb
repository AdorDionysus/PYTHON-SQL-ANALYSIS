```python
import pandas as pd
```

```python
df = pd.read_csv('orders.csv',parse_dates=['Order Date'],na_values=['unknown','Not Available'])

df['Ship Mode'].unique()
```

```python
#renaming columns to lowercase and change spacing into underscore to avoid error when we use sql later
df = df.rename(
    columns={
        'Order Id': 'order_id',
        'Order Date':'order_date'
        ,'Ship Mode':"ship_mode",
        'Segment':'segment',
        'Country':'country',
        'City':'city',
        'State': 'state',
       'Postal Code':'postal_code',
        'Region':'region',
        'Category':'category',
        'Sub Category':'sub_category',
        'Product Id':'product_id',
       'cost price':'cost_price',
        'List Price' : 'list_price',
        'Quantity':'quantity',
        'Discount Percent':'discount_percent'
    }
).copy()
df
```

```python
# CREATING DISCOUNT,SALES PRICE,PROFIT AND TOTAL PROFIT COLUMN



df['discount'] = df['list_price'] * df['discount_percent'] * .01
df['sale_price'] = df['list_price'] - df['discount']

# PROFIT COLUMN
df['profit'] = df['sale_price'] - df['cost_price']


#  TOTAL PROFIT COLUMNS (TOTAL PROFIT COLUMNS IS MULTIPLIED BY QUANTITY UNLIKE TO PROFIT COLUMNS)
df['total_profit'] = df['profit'] * df['quantity']
df
```

```python
# DROPPING COST PRICE, DISCOUNT PRICE AND LIST PRICE


df = df.drop(columns=['cost_price','list_price','discount_percent'])
df
```

```python
df['totalrev'] = df['quantity'] * df['sale_price']
grouped = df.groupby('product_id')
sum = grouped['totalrev'].mean
sum.sort_values(ascending=False)
```
